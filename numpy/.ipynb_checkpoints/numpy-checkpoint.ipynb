```python
import numpy as np

# 1D array
arr1 = np.array([1,2,3,4,5,6])

# 2D array

arr2 = np.array([
                    [1,2,3],
                    [4,5,6]
                ])


#3D array
arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2 

    #2nd floor

     [[1,2,3], # row1
      [4,5,6]  ], #row2  


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2  

    
])
```

### Array Attributes

```python
# dimension 
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim) 

print("="*30)

#shape 
print(arr1.shape)
print(arr2.shape)
print(arr3.shape) 

print("="*30)

#data type 
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

print("="*30)

#size 
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

### Upcasted

```python
# upcasted float
arr = np.array([1,2,3.2])
print(arr.dtype)

# upcasted string
arr = np.array([1,2,3.4,"hello"])
print(arr.dtype)

arr = np.array([1,True,3.25,'hello',]) 
print(arr.dtype)
```

### Selecting data type for an array

```python
arr = np.array([1,2,3], dtype=np.int16) 
print(arr.dtype)

arr = arr.astype(np.int32)
print(arr.dtype)

#error
arr = np.array([1,2,3.2,'hello'],dtype=np.float64) 
```

### ndarray creation from existing data


```python
lst = [10,20,30,40 , 40.5] 
arr = np.array(lst, dtype=np.int32)

print(arr)
print(type(arr))
print(arr.dtype)
```

```python
mixed_lst = [10, True , 'Hello' ]

arr = np.array(mixed_lst) 

print(arr) 
print(type(arr)) 
print(arr.dtype) 
```

```python
matrix = [ [1,2,3], # row1
           [4,5,6] #row2
                ]

arr = np.array(matrix) 

print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
```

```python
# tuple

tpl = (10,20,30)
arr = np.array(tpl, dtype=np.int32)
print(arr.dtype)
arr = arr.astype(np.int16)
print(arr.dtype)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
# dictionary 
dc = {'a':10, 'b':20, 'c':30}

keys = dc.keys()
values = dc.values()
items = dc.items()
```

```python
arr = np.array(list(keys))
print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
#values to ndarray
arr = np.array(list(values))
print(arr) 
arr = arr.astype(np.int16)
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
arr = np.array(list(items))
print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
arr = np.zeros_like(arr3)
print(arr)
print(arr.shape)
```

```python
arr = np.ones((3,4,3),dtype=np.int8)
print(arr)
print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
arr = np.ones_like(arr3) ## arr3 shape same 
print(arr)
print(arr.shape)
```

```python
#np.empty
arr = np.empty((4,3),dtype=np.int8)
print(arr)
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

### array creation with random values


```python
arr = np.random.rand(3,3) # values are from 0 -> 1
print(arr)
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.random.randint(start,end, shape) 
arr = np.random.randint(1,10,(2,3))
print(arr)
print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.random.uniform(start,end, shape) 
arr = np.random.uniform(1,10,(3,3,4))
print(arr)
```

```python
# np.arange(first , end , step ) -> similar to range function 
arr = np.arange(1,10,1)
print(arr)
print("-"*20)
mat = arr.reshape(3,3)
print(mat)
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.linspace(start , end(included) , koita) 
arr = np.linspace(1,10,15)
print(arr)
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

### indexing and slicing

```python
#indexing 
print(arr1)
print("-"*20)
print(arr2)

print("-"*20)

arr1[2] = 100
print(arr1)

print("-"*20)

arr2[0][2] = 100
print(arr2)
```

```python
#slicing 
#1d = arr[start:end:step] 

arr_mod = arr1[1:4].copy()

print(arr1)

arr_mod[1] = 3

print(arr_mod)
```

```python
#2d slicing 
# arr[ row_start : row_end : step , column_start : column_end : step ]

print(arr2)
print("-"*20)

# getting a row
row_0 = arr2[0:1,]
row_1 = arr2[1:2,]

print(row_1)
print(row_0)

print("-"*20)

#getting a column 

col_0 = arr2[::,0:1]
col_1 = arr2[::,1:2]
print(col_0)
print("-"*20)
print(col_1)

print("-"*20)
#getting a portion 
portion = arr2[::,0:2]
print(portion)
```
