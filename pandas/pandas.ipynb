```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
print(df)  
```

```python
df_filled1 = df.fillna({'A':0, 'B': 99, 'C':'Unknown'})
df_filled1
```

```python
df_ff = df.copy()
df_ff["B"] = df_ff["B"].ffill()
print(df_ff)
```

```python
for column in df.columns:
    if df[column].dtype == "object":
        df[column].fillna("Unknown", inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)
df
```

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

# Drop rows where column A has missing values
df_a = df.dropna(subset=["A"])

print(df_a)
```

```python
df_clean = df.dropna(thresh=2)
print(df_clean)
```

```python
df_clean = df.dropna(subset=["A","B"], how="all")
print(df_clean)
```
