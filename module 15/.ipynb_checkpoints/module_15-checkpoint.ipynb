```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 2D line plot

```python
hours = [2 , 3 , 4, 1 , 5 , 7 ,3] 
days = [1 ,2 , 3 ,4 ,5 ,6 ,7]

plt.plot(days, hours)

plt.show()
```

```python
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Trend os a student study time over a week")
plt.plot(days, hours)
plt.show()
```

```python
hours_1 = [2 , 3 , 4, 1 , 5 , 7 , 3] 
hours_2 = [ 1 , 4 , 3 , 5 ,7 ,2 ,1]
days = [1 ,2 , 3 ,4 ,5 ,6 ,7] 

plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Trend os a student study time over a week")

plt.plot(days, hours_1)
plt.plot(days, hours_2)
```
