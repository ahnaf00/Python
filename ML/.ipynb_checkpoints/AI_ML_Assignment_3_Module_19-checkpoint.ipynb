```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVR, SVC
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

## TODO A1: Dataset Loading and Understanding (5 Marks)

```python
# TODO: Load regression dataset
df_reg = pd.read_csv("insurance.csv")
print(df_reg.shape)
print(df_reg.info())
```

target variable = "charges".
The dataset provides individual medical cost 

## TODO A2: Exploratory Data Analysis (EDA) (5 Marks)
**You must:**
- Show descriptive statistics
- Plot correlation heatmap
- Plot at least one feature vs target

**Marks:** Plots (3), Interpretation (2)


```python
# TODO: Perform EDA
print(df_reg.describe().T)
print(df_reg.isnull().sum())
```

The dataset shows 0 missing values, average medical cost 13270.42, median 9382.033. The average bmi 30.66 which fall into obese. The top 25% people have BMI over 26.29

```python
numerical_cols = ["age","bmi","children","charges"]
categorical_col = ["sex","smoker","region"]
plt.figure(figsize=(8,6))
sns.heatmap(df_reg[numerical_cols].corr(),annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df_encoded = pd.get_dummies(df_reg,drop_first=True)
df_encoded = df_encoded.astype(int)
df_encoded.head()
```

## TODO A3: Feature Preparation (5 Marks)
**You must:**
- Train-test split
- Scaling (if applicable)

**Marks:** Correct split (3), Explanation (2)


```python
# TODO: Prepare features
X_reg = df_encoded.drop("charges",axis=1)
y_reg = df_encoded["charges"]
X_train_r,X_test_r,y_train_r,y_test_r = train_test_split(X_reg,y_reg,test_size=0.25,random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_r)
X_test_scaled = scaler.transform(X_test_r)
```

Here we split the data for training and testing. Then using StandardScaler() to normalize all features to have mean 0 and std 1. Scaling ensures features with different ranges are on the same scale

## TODO A4: Multiple Linear Regression (8 Marks)
**You must:**
- Train model
- Evaluate using RMSE, MAE, R²

**Marks:** Model (4), Metrics (4)


```python
# TODO: Multiple Linear Regression
model = LinearRegression()
model.fit(X_train_scaled,y_train_r)
y_pred = model.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test_r,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_r,y_pred)

print("mse: ",mse)
print("rmse: ",rmse)
print("r2: ",r2)
```

## TODO A5: Polynomial Regression (7 Marks)
**You must:**
- Apply polynomial features
- Compare with linear regression

**Marks:** Implementation (4), Comparison (3)


```python
# TODO: Polynomial Regression
poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train_scaled)
X_poly_test = poly.transform(X_test_scaled)

p_model = LinearRegression()
p_model.fit(X_poly_train,y_train_r)

y_pred_poly = p_model.predict(X_poly_test)
mse_p = mean_squared_error(y_test_r,y_pred_poly)
r2_p = r2_score(y_test_r,y_pred_poly)

print("mse_p: ",mse_p)
print("r2_p: ",r2_p)
```

Linear regression works with straight line where polynomial regression works with non-linear patterns. Polynomial regression usually get lower MSE and higher R2 values. Polynomial regression performs better when relationships are non-linear

## TODO A6: Support Vector Regression (7 Marks)
**You must:**
- Train SVR (RBF kernel)
- Evaluate performance
- Justify kernel choice

**Marks:** Model (4), Justification (3)


```python
# TODO: Support Vector Regression
from sklearn.svm import SVR

svr = SVR(kernel='rbf')
svr.fit(X_train_scaled,y_train_r)

y_pred_svr = svr.predict(X_test_scaled)

mse_svr = mean_squared_error(y_test_r,y_pred_svr)
r2_svr = r2_score(y_test_r,y_pred_svr)

print("mse_svr: ",mse_svr)
print("r2_svr: ",r2_svr)
```

RBF handles non-linera data better than others. It maps data into higher dimensional space which makes its performance better

## TODO A7: Random Forest Regressor (8 Marks)
**You must:**
- Train Random Forest
- Evaluate model

**Marks:** Model (4), Evaluation (4)


```python
# TODO: Random Forest Regressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
rf.fit(X_train_r,y_train_r)

y_pred_rf = rf.predict(X_test_r)

mse_rf = mean_squared_error(y_test_r,y_pred_rf)
r2_rf = r2_score(y_test_r,y_pred_rf)

print("mse_rf: ",mse_rf)
print("r2_rf: ",r2_rf)
```

Random forest reduces overfitting using ensembling learning. provides best performance among regression models

# 🟠 PART B: Classification Task (45 Marks)
You will work with the provided **classification dataset**.


## TODO B1: Dataset Understanding (5 Marks)
**You must:**
- Load dataset
- Show class distribution
- Comment on imbalance

**Marks:** Analysis (3), Explanation (2)


```python
# TODO: Load classification dataset
df_loan = pd.read_csv("loan_data.csv")

display(df_loan.head())
print(df_loan.info())
```

The plot reveals if the dataset is imbalanced which affects model choice and evaluation

## TODO B2: Exploratory Data Analysis (5 Marks)
**You must:**
- Plot class counts
- Plot at least one feature vs class

**Marks:** Plots (5),


```python
# TODO: Classification EDA
sns.countplot(x='loan_status',data=df_loan)
plt.title("Loan approval distribution")
plt.show()

numerical_cols_loan = ["person_age","person_income","person_emp_exp","loan_amnt","loan_int_rate","loan_percent_income","cb_person_cred_hist_length","credit_score","loan_status"]
categorical_col_loan = ["person_gender","person_education","person_home_ownership","loan_intent","previous_loan_defaults_on_file"]

plt.figure(figsize=(10,8))
sns.heatmap(df_loan[numerical_cols_loan].corr(),annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## TODO B3: Feature Preparation (5 Marks)
**You must:**
- Encoding (if needed)
- Scaling
- Train-test split with stratification




```python
# TODO: Prepare classification features
from sklearn.preprocessing import LabelEncoder
df_loan.isna().sum()
```

```python
le = LabelEncoder()
for col in categorical_col_loan:
    df_loan[col] = le.fit_transform(df_loan[col])

X_loan = df_loan.drop("loan_status",axis=1)
y_loan = df_loan["loan_status"]

scaler_loan = StandardScaler()
X_scaled_loan = scaler_loan.fit_transform(X_loan)

print(X_scaled_loan)
```

```python

X_train_l,X_test_l,y_train_l,y_test_l = train_test_split(
    X_scaled_loan,
    y_loan,
    test_size=0.2,
    random_state=42,
    stratify=y_loan
)

print(f"TraiN size: {X_train_l.shape[0]}")
print(f"Test Size: {X_test_l.shape[0]}")
```

## TODO B4: Logistic Regression with L2 (8 Marks)
**You must:**
- Train Logistic Regression
- Evaluate using Accuracy, Precision, Recall
- Show Confusion Matrix

**Marks:** Model (4), Metrics (4)


```python
# TODO: Logistic Regression
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train_l,y_train_l)

y_pred_log = log_reg.predict(X_test_l)
print("Accuracy: ",accuracy_score(y_test_l,y_pred_log))
print(classification_report(y_test_l,y_pred_log))
```

```python
cm_reg = confusion_matrix(y_test_l,y_pred_log)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_reg, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix: Logistic Regression')
plt.xlabel('Predicted Label (0=Reject, 1=Approve)')
plt.ylabel('Actual Label')
plt.show()
```

## TODO B5: Support Vector Machine (7 Marks)
**You must:**
- Train SVM (Linear or RBF)
- Evaluate

**Marks:** Model (4), Evaluation (3)


```python
# TODO: SVM Classification
svm_rbf = SVC(kernel='rbf',C=1.0,gamma='scale',random_state=42)
svm_rbf.fit(X_scaled_loan,y_loan)

y_pred_loan = svm_rbf.predict(X_test_l)

print("Accuracy: ",accuracy_score(y_test_l,y_pred_loan))
print(classification_report(y_test_l,y_pred_loan))
```

## TODO B6: Naive Bayes (5 Marks)
**You must:**
- Choose correct NB variant



```python
# TODO: Naive Bayes
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train_l,y_train_l)

y_pred_gnb = gnb.predict(X_test_l)

print("accuracy: ",accuracy_score(y_test_l,y_pred_gnb))
print(classification_report(y_test_l,y_pred_gnb))
```

## TODO B7: K-Nearest Neighbors (5 Marks)
**You must:**
- Try multiple K values
- Plot accuracy vs K

**Marks:** Experiment (3), Plot & choice (2)


```python
# TODO: KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_l,y_train_l)

y_pred_knn = knn.predict(X_test_l)
print("Accuracy: ",accuracy_score(y_test_l,y_pred_knn))
```

```python
k_values = range(1, 31)
accs = []

for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_l,y_train_l)
    accs.append(knn_temp.score(X_train_l,y_train_l))

plt.figure(figsize=(10, 5))
plt.plot(k_values, accs, marker='o')
plt.title('Accuracy vs. K Value')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

best_k = k_values[np.argmax(accs)]
print(f"Best k found: {best_k}")
```

## TODO B8: Random Forest Classifier (5 Marks)
**You must:**
- Train model
- Show feature importance

**Marks:** Model (3), Interpretation (2)


```python
# TODO: Random Forest Classifier
rf_baseline = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_baseline.fit(X_train_l,y_train_l)
y_pred_baseline = rf_baseline.predict(X_test_l)

print("Accuracy: ",accuracy_score(y_test_l,y_pred_baseline))
```

```python
importances = pd.Series(
    rf_baseline.feature_importances_,
    index=X_loan.columns
).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
importances.plot(kind='bar', color='teal')
plt.title("Feature Importance (Random Forest)")
plt.ylabel("Importance Score")
plt.show()
```
