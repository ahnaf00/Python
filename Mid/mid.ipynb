```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

```python
# 1 

ages = np.array(ages)
print(ages)
print(ages.dtype)
print(ages.ndim)
print(ages.shape)
print(ages.size)

print("-"*30)



salaries = np.array(salaries)
print(salaries)
print(salaries.dtype)
print(salaries.ndim)
print(salaries.shape)
print(salaries.size)
```

```python
# 2
print(salaries)
highest_salary = np.max(salaries)
lowest_salaries = np.min(salaries)
avg_salary = np.average(salaries)

print("Highest salary ",highest_salary)
print("Lowest salary: ",lowest_salaries)
print("Average salary: ",avg_salary)

average_age = np.average(ages)
print("Average age: ",average_age)
```

```python
# 3
print(ages)
age_greater_than_30 = ages[ages>30]
print(age_greater_than_30)
age_greater_than_30_count = np.count_nonzero(ages>30)
print(age_greater_than_30_count)
```

```python
# 4
new_arra = ages+5
print(new_arra)
```

```python
# 5
total_salary = np.sum(salaries)
print("Total salary: ",total_salary)

diff = np.max(salaries) - np.min(salaries)
print("Difference between highest and lowest salary: ",diff)
```
